## Standard Scalar

```python
# Calculates mean & std dev of each col, mean=0, std dev=1
```

```python
import pandas as pd
```

```python
df = pd.read_csv('D:/Kaggle Datasets/IRIS.csv')
```

```python
df
```

```python
X = df.drop('species', axis = 1)
Y = df['species']
```

```python
X.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
# transfroming non-scaled data into scaled data
scaler.fit_transform(X)
# fit - it means learns all the parameters like mean & std dev of each column of dataset
# transform - means start subtracting mean from each value in the column and then divide it with std dev
```

```python
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled.shape
```

```python
X_scaled[:,2]  # all the rows and second index column
```

```python
X_scaled[:,2].mean()  # approx. 0
```

```python
X_scaled[:,2].std()  # approx. 1
```
